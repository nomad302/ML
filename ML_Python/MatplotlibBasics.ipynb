```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x=np.arange(4)
x=x**2.7
x2=x**3
plt.plot(x)
plt.plot(x2,color="green")
plt.show()
```

```python
plt.scatter(x,x*2,marker="^",color="orange",label="stock price")
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock price prediction")
plt.show()
```
