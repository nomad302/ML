# Generating Regression Data

```python
import numpy as np
import matplotlib.pyplot as plt

u=0
std=1
x= (np.random.randn(300)*std + u) # randn return a normal distribution, standard normal distribtion has zero mean 
noise=np.random.randn(300)
# and unit variance
#plt.hist(x)
#plt.show()
```

```python
y=3*x +2 +noise   # EQUATION of line y=mx+c
print(x.shape,y.shape)
```

```python
plt.scatter(x,y,color="orange")
plt.show()
```

# Generating Classification data

### Lets us generate some examples for Binary Classification

```python
# for this we are going to use multi-variate normal distribution, 
# normal distribution is used when there is single feature and
# Multi-Variate distribution is used where there is more than one feature 
# in training data points.
```

```python
# for multi-variate normal distribtion we need to specify mean of distribution u1 and u2
# and we also need to specify co-Variance
```

```python
u1=np.array([2,4])
cov_1=np.array([[2,-0.8],
                [-0.8,1]]) 

u2=np.array([-1,-3])
cov_2=np.array([[1,0],
                [0,2]]) 

x1 = np.random.multivariate_normal(u1,cov_1,300)
print(x1.shape)

x2 = np.random.multivariate_normal(u2,cov_2,400)
print(x2.shape)
```

```python
plt.scatter(x1[:,0],x1[:,1],label="Apple") # from all row 0th row goes on x-axis and 1th goes on y-axis
plt.scatter(x2[:,0],x2[:,1],marker="^",label="Oranges")
plt.legend()
plt.title("Apples and Oranges")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.show()
```

```python
y = np.zeros(700)
y[:300]=1 # as fisrt 300 rows are apples
print()
```

```python
x = np.vstack((x1,x2)) # vertical statck
print(x.shape)
```

```python
#plt.scatter(x[:,0],x[:,1])
for i in range(700):
    if y[i] == 1:
        plt.scatter(x[i,0],x[i,1],color="Green")
    else:
        plt.scatter(x[i,0],x[i,1],color="Red")
        
        
plt.show()
```

## shortcut

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```
