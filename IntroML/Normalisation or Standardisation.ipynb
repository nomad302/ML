### Data Preproccesing 

Normalisation means where mean of data should be zero and standard deviation  should be one.
mean (u) = 0
std = 1

how to do Normalisation ?

Normalisation = x - u / std

where x is data points.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfx = pd.read_csv('/home/shaggy/CodingBlocks/NoteBooks/Datasets/LinearX.csv')
```

```python
dfy = pd.read_csv('/home/shaggy/CodingBlocks/NoteBooks/Datasets/LinearY.csv')
```

```python
x = dfx.values
y = dfy.values

#Normmalisation
x2 =(x-x.mean())/x.std()
plt.scatter(x,y,label='Raw data')
plt.scatter(x2,y,color='orange',label='Normalised data')
plt.legend()
plt.show()
```
