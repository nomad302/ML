### Generating Datasets

```python
from sklearn.datasets import make_regression,make_classification,make_blobs,make_moons
from matplotlib import pyplot as plt
```

```python
x,y =make_regression?
```

```python
x,y =make_regression(n_samples=100,n_features=1,n_informative=1,noise=10.8)
print(x.shape,y.shape)
```

```python
plt.scatter(x,y)
```

### Generating classification Dataset

```python
xc,yc=make_classification(n_samples=400,n_features=2,n_classes=3,n_clusters_per_class=1,n_informative=2,n_redundant=0,random_state=5)
print(xc.shape,yc.shape)
plt.scatter(xc[:,0],xc[:,1],c=yc)
plt.show()
```

### Blobs Datasets

```python
xb,yb=make_blobs?
```

```python
xb,yb=make_blobs(n_samples=100,n_features=2,centers=5,random_state=10)
```

```python
plt.scatter(xb[:,0],xb[:,1],c=yb)
plt.show()
```

### Moons Datasets

```python
xm,ym=make_moons(n_samples=100,noise=0.5,random_state=5)
plt.scatter(xm[:,0],xm[:,1],c=ym)
plt.show()
```

### Subplots

```python
plt.figure(figsize=(12,12))
plt.subplot(221)
plt.title("Regresssion")
plt.scatter(x,y)


plt.subplot(222)
plt.title("Classification")
plt.scatter(xc[:,0],xc[:,1],c=yc)


plt.subplot(223)
plt.title("Blob")
plt.scatter(xb[:,0],xb[:,1],c=yb)

plt.subplot(224)
plt.title("Moon")
plt.scatter(xm[:,0],xm[:,1],c=ym)

plt.show()
```
