### Normal Distribution

Normal Distribution is parameterized by two things :
1. u - mean of distribution
2. sigma - standard deviation of distribution


### HIstoGram

```python
# Generate Normal Distribution using Numpy function (1-D)
```

```python
import numpy as np
import matplotlib.pyplot as plt
u = 5   # u is mean 
sigma = 2 # sigma defines spread of curve
# such distribution is also called standard normal distribution
vals = u + sigma*(np.random.randn(200))
print(vals.shape)
plt.hist(vals,10)
plt.show()
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

### Multivariate Normal Distribution

```python
#BiVariate

mean = np.array([0.0,0.0])   # distribtion is centered around origin
cov = np.array([[1,0],[0,3]]) 
dist = np.random.multivariate_normal(mean,cov,500) 


mean2 = np.array([5.0,6.0])   
cov2 = np.array([[1.3,0.2],[0.2,1.3]]) 
dist2 = np.random.multivariate_normal(mean2,cov2,500) 

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
