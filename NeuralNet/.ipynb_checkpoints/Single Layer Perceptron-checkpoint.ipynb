# Model and Helper Function

```python
def sigmoid(z):
    return (1.0)/(1+np.exp(-z))

sigmoid(12.504)
```

# Implement Perceptron Learning Algorithm
- Learn the weights
- Reduce the loss
- make the Predictions

```python
def predict(X,weights):
    """X is mx(n+1) matrix ,W --> (nx1) vector"""
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions
```

```python
 def loss(X,Y,weights):
    """Binary cross Entropy"""
    Y_ = predict(X,weights)
    cost = np.mean(-Y*np.log(Y_) - (1-Y)*np.log(1-Y_))
    return cost
        
```

```python
a = np.array([1,2,3])
np.mean(a)
```

```python
def update(X,Y,weights,learning_rate):
    """Performing weight updates for 1 epoch"""
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_ - Y)
    
    m = X.shape[0]
    weights = weights - learning_rate* dw/(float(m))
    return weights
```

```python
def train(X,Y,learning_rate=0.5,maxEpochs=100):
    # Modify the input to handle the bias term
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    #init Weights 0
    weights = np.zeros(X.shape[1])# n+1 entries
    
    # Iterate over all epochs and make updates
    
    for epoch in range(maxEpochs):
        weights = update(X,Y,weights,learning_rate)
        
        if epoch%20 ==0 :
            l = loss(X,Y,weights)
            print("Epoch %d Loss %.4f"%(epoch,l))
            
            
    return weights
    
```

```python
weights = train(X,Y)
```

- make Predictions
- Visualise Decision surface
- Linear vs Non-Linear Classification

```python
def predictions(x_Test,weights,labels=True):
    if x_Test.shape[1] != weights.shape[0]:
        ones = np.ones((x_Test.shape[0],1))
        x_Test = np.hstack((ones,x_Test))
        
    probs = predict(x_Test,weights)
    
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs > 0.5] = 1
        return labels
```
