```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfx = pd.read_csv('/home/shaggy/Machine_Learning/KNN/xdata.csv')
dfy = pd.read_csv('/home/shaggy/Machine_Learning/KNN/ydata.csv')
```

```python
X = dfx.values
Y = dfy.values

X = X[:,1:] # drop the first column
Y = Y[:,1:].reshape((-1,))

print(X.shape)
print(Y.shape)
```

```python
query_x = np.array([0,0])
plt.scatter(query_x[0],query_x[1],color='red')
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals =  sorted(vals)
    # nearest or first K points
    vals = vals[:k]
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1],return_counts=True)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
pred = knn(X,Y,query_x)
print(pred)
```
